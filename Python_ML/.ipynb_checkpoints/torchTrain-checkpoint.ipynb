```python
import torch
import torchfcn
import matplotlib.pyplot as plt
import os

# 参数初始化
def get_parameters(model, bias=False):
    import torch.nn as nn
    modules_skipped = (
        nn.ReLU,
        nn.MaxPool2d,
        nn.Dropout2d,
        nn.Sequential,
        torchfcn.models.FCN32s,
        torchfcn.models.FCN16s,
        torchfcn.models.FCN8s,
    ) 
    for m in model.modules():
        if isinstance(m, nn.Conv2d):
            if bias:
                yield m.bias
            else:
                yield m.weight
        elif isinstance(m, nn.ConvTranspose2d):
            if bias:
                assert m.bias is None
        elif isinstance(m, modules_skipped):
            continue
        else:
            raise ValueError('Unexpected module: %s' % str(m))
```

```python
# 创建自己的数据集类
class MyDataset(torch.utils.data.Dataset):
    def __init__(self, X, Y):
        super(MyDataset, self).__init__()
        # X, Y 类初始化传入数据
        # X 维度 （图片张数，3个RGB通道，图片长，图片宽）
        # Y 维度 同上
        
        # *****************下面自己写**************************
        self.data = None              # 创建一个空列表
        self.length = None            # 数据的个数（图片的张数），从X的维度中获得
        for i in range(self.length):
            self.data.append(None)    # 将（X_i，y_i）加入data列表，即一组（数据-标签）对加入列表
        
    def __getitem__(self, i):
        return None                   # 返回第i个（数据-标签）组
    
    def __len__(self):
        return None                   # 返回数据的个数 
        #******************上面自己写**************************
```

```python
def load_data():
    
    #*****************下面自己写**************************
    datapath = None                    # img-train 文件夹的路径
    labelpath = None                   # img-label 文件夹的路径
    #******************上面自己写**************************
    
    length = len(os.listdir(datapath)) # 获取图片数量
    X = []
    Y = []
    for i in range(1, length + 1):
        
        #*****************下面自己写**************************
        data_name = None               # img-train中第i图片名，e.g. train_1.bmp
        label_name = None              # label图片名
        
        data_img = plt.imread(None)    # 组合路径和文件名
        label_img = plt.imread(None)
        
        X.append(None)                 # 将data_img加入X中
        Y.append(None)                 # 将label_img加入Y中
        
        #*****************上面自己写**************************
        
    #*****************下面自己写**************************
    X = None                           # 转换为numpy数组
    Y = None                           # 转换为numpy数组 
        
    X = X.transpose(None)              # 转置X，将维度变为（图片张数，3个RGB通道，图片长，图片宽）
    Y = Y.transpose(None)              # 转置Y，将维度变为（图片张数，3个RGB通道，图片长，图片宽）
    
    train = MyDataset(None, None) # 传入X和Y
    #******************上面自己写**************************
    
    train_loader = torch.utils.data.DataLoader(train, shuffle = True)
    return train_loader
```

```python
def main(lr = None, momentum = None, weight_decay = None, max_iteration = None): # 填入example中的默认值
    # 导入数据
    train_loader = load_data()
    
    # 下面一行自己写
    model = torchfcn.models.FCN32s(n_class = None) # 创建网络模型，输入类别数，即label图像中颜色的种类
    
    # 调用显卡
    model = model.cuda()

    # 设置优化方法
    optim = torch.optim.SGD(
            [
                {'params': get_parameters(model, bias = False)},
                {'params': get_parameters(model, bias = True),
                 'lr': lr * 2, 'weight_decay': 0},
            ],
            lr = lr,
            momentum = momentum,
            weight_decay = weight_decay)
    
    # 设置训练器
    trainer = torchfcn.Trainer(
        cuda = True,
        # *****************自己写*************
        model = None,                 # 网络模型
        optimizer = None,             # 优化方法
        train_loader = None,          # 数据加载器
        max_iter = None,              # 迭代次数
        # *************自己写*****************
        val_loader = train_loader,
        out = '/home/tongxueqing/tong/out.log',
    )
    
    # 开始训练
    trainer.epoch = 0
    trainer.iteration = 0
    trainer.train()
    
if __name__ == '__main__':
    main()
```
